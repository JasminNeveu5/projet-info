# Quel est le pilote avec le plus d'acccident

```python
import pandas as pd
from pathlib import Path
import sys
from options.config import DATA_DIR
PROJECT_ROOT = Path.cwd().parent  # Assuming you open Jupyter inside 'projet_info/questions/'
sys.path.append(str(PROJECT_ROOT))

status = pd.read_csv(f"{DATA_DIR}/status.csv")
races = pd.read_csv("~/Documents/1A-ENSAI/projet_info/data/races.csv")
results = pd.read_csv("~/Documents/1A-ENSAI/projet_info/data/results.csv")
drivers = pd.read_csv("~/Documents/1A-ENSAI/projet_info/data/drivers.csv")
status.head()
```

```python
results_accident = results.query("statusId == 3").groupby("driverId").agg(nombre_accident = ("statusId","sum")).sort_values("nombre_accident", ascending=False).merge(drivers,on = "driverId")[["driverId","forename","surname","nombre_accident"]]
results_accident.head()
```

Là on a eu le nombre d'accidents par pilote mais bon ça ne permet pas trop de comparer, le nombre de courses augementent forcément la probabilité d'avoir un accident. On va donc faire calculer pour chaque pilotes le ration nb courses/nb accidents

```python
results_total = results.groupby("driverId").size().reset_index(name="count")
results_total = pd.merge(results_total, results_accident,on = "driverId")
results_total
```

```python
results_total["ratio"]  = 100*results_total["nombre_accident"] / results_total["count"]
results_total.sort_values("ratio",ascending=False)
```
