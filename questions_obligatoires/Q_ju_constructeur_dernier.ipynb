# Quel constructeur finit le plus souvent dernier des courses ?

Cette question traite des coureur.euses qui finissent la course et n'aborde pas celles et ceux qui ne finissent pas la course.

* Traitement par python.
* Jointure entre constructors.csv et constructor_standings.csv via la clé constructorId
* Trasnformation du dataframe en liste de listes pour traiter la question en python

```python
import pandas as pd
import matplotlib.pyplot as plt

constructor_standings = pd.read_csv(
                'C:/DEVOIRS/ENSAI1A/projet-info/data/constructor_standings.csv'
                )
constructors = pd.read_csv('C:/DEVOIRS/ENSAI1A/projet-info/data/constructors.csv')

jointure = pd.merge(constructor_standings, constructors, on='constructorId', how='left')

jointure_liste = jointure.values.tolist()
jointure
```

```python
jointure_liste
```

Avoir la dernière place de chaque course

```python
last_position_per_race = [-1 for i in range(1144+1)]  # 1144 courses dans la table constructor_standings

for line in jointure_liste:
    if isinstance(last_position_per_race[line[2]], int):
        if int(line[5]) > int(last_position_per_race[line[2]]):
            last_position_per_race[line[2]] = line[5]
```

```python
last_positions_per_constructor = [0 for i in range(215+1)]  # 215 constructeurs dans la table constructor_standings
print(last_positions_per_constructor)
for line in jointure_liste:
    raceId = line[2]
    position = line[5]
    constructorId = int(line[3])
    if last_position_per_race[raceId] == position and constructorId < 216:
        last_positions_per_constructor[constructorId] += 1
```
